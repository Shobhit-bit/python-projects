```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"D:\vs code\Projects\MachineLearning\IRIS.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()
df['species']=le.fit_transform(df['species'])
```

```python
x = df.drop(columns="species")
y = df['species']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
y.tail()
```

With Logistic Regrestion

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_LR = LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
y_pred_LR = model_LR.predict(x_test)
```

```python
accuracy_LR = accuracy_score(y_test,y_pred_LR)
print(accuracy_LR)
```

```python
tem_LR = model_LR.predict([[51,8.9,4.5,3.4]])
flowers = ["setosa","virginica","versicolor"]
print(flowers[int(tem_LR)])
```

With SVM algo

```python
from sklearn.svm import SVC
```

```python
model_SVC = SVC()
model_SVC.fit(x_train,y_train)
```

```python
y_pred_SVC= model_SVC.predict(x_test)
accuracy_SVC = accuracy_score(y_test,y_pred_SVC)
print(accuracy_SVC)
```

```python
tem_SVC = model_SVC.predict([[51,8.9,4.5,3.4]])
flowers = ["setosa","virginica","versicolor"]
print(flowers[int(tem_SVC)])
```

With Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DTC = DecisionTreeClassifier()
model_DTC.fit(x_train,y_train)
```

```python
y_pred_DTC = model_DTC.predict(x_test)
accuracy_DTC = accuracy_score(y_test,y_pred_DTC)
print(accuracy_DTC)
```

K-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNC = KNeighborsClassifier()
model_KNC.fit(x_train,y_train)
```

```python
y_pred_KNC = model_KNC.predict(x_test)
accuracy_KNC = accuracy_score(y_test,y_pred_KNC)
print(accuracy_KNC)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RFC = RandomForestClassifier()
model_RFC.fit(x_train,y_train)
```

```python
y_pred_RFC = model_KNC.predict(x_test)
accuracy_RFC = accuracy_score(y_test,y_pred_RFC)
print(accuracy_RFC)
```
