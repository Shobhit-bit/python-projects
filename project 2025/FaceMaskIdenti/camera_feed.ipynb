```python
import cv2
```

```python
import pickle
model = pickle.load(open("D:\model.pkl","rb"))
```

```python
def detect_face(img):
    y_pred = model.predict(img.reshape(1,224,224,3))
    return y_pred[0][0]
```

```python
cap = cv2.VideoCapture(0)
```

```python
def writ (img,text,pos,bg_color):
    text_size = cv2.getTextSize(text,cv2.FONT_HERSHEY_PLAIN,1,cv2.FILLED)
    end_x = pos[0] + text_size[0][0] + 2
    end_y = pos[1] + text_size[0][1] - 2
    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_PLAIN,1,(0,0,0),1,cv2.LINE_AA)
```

```python
while True:
  ret,frame = cap.read()
  img = cv2.resize(frame,(224,224))
  y_pred = detect_face(img)
  if y_pred <= 0.1:
    writ(frame,"Mask",(10,10),(0,255,0))
  else:
    writ(frame,"No Mask",(10,10),(0,0,255))
  cv2.imshow("Window",frame)
  if cv2.waitKey(1) & 0xFF == ord('x'):
    break
```

```python
cv2.destroyAllWindows()
```
