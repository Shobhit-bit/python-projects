```python
import numpy as np
import pandas as pd
```

Importing Database

```python
movie= pd.read_csv(r'D:\vs code\Projects\MachineLearning\Movie_Recommend\tmdb_5000_movies.csv')
credit = pd.read_csv(r'D:\vs code\Projects\MachineLearning\Movie_Recommend\tmdb_5000_credits.csv')
```

```python
movie.head()
```

```python
credit.head()
```

Merging On Title

```python
movie = movie.merge(credit,on = 'title')
movie.head()
```

Droping Duplicate Column

```python
movie = movie.drop(columns= ['movie_id'])
movie.head()
```

Creating new database with relevant columns

```python
recommendation = movie[['genres','id','keywords','title','overview','cast','crew']]
recommendation.head()
```

```python
recommendation.isnull().sum()
```

```python
recommendation.dropna(inplace=True)
```

Taking out genre from Dictionary and removing spaces

```python
import ast
def extraction(object):
    object = ast.literal_eval(object)
    genre = []
    for i in object:
        genre.append(i["name"])
    return genre
```

```python
recommendation["genres"] = recommendation["genres"].apply(extraction)
recommendation["genres"] = recommendation["genres"].apply(lambda x:[i.replace(" ","") for i in x])
recommendation["genres"].head()
```

Same for Keywords

```python
recommendation["keywords"] = recommendation["keywords"].apply(extraction)
recommendation["keywords"] = recommendation["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

Extraction of cast from dictionary and removal of spaces

```python
def extraction(object):
    object = ast.literal_eval(object)
    cast = []
    count = 0
    for i in object:
        if count < 3:
            cast.append(i["name"])
            count += 1
            continue
        break
    return cast
```

```python
recommendation["cast"] = recommendation["cast"].apply(extraction)
recommendation["cast"] = recommendation["cast"].apply(lambda x:[i.replace(" ","") for i in x])
```

Extraction Of crew

```python
recommendation["crew"] = recommendation["crew"].apply(extraction)
recommendation["crew"] = recommendation["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

Cleaning Overview for stop words

```python
from nltk.corpus import stopwords
stop_words = list(stopwords.words("english"))
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)
recommendation["overview"] = recommendation["overview"].apply(remove_stopwords)
```

Converting Overview into list for concatination

```python
recommendation["overview"] = recommendation["overview"].apply(lambda x: x.split())
```

```python
recommendation.head()
```

merging overview,cast,crew,genre,keywords to create a new column

```python
recommendation['tags'] = recommendation['genres'] + recommendation['keywords'] + recommendation['cast'] + recommendation['crew'] + recommendation['overview']
recommendation.drop(columns = ['genres', 'keywords', 'cast', 'crew', 'overview'], inplace=True)
```

### Cleaning Tags column

```python
recommendation['tags'] = recommendation['tags'].apply(lambda x: " ".join([str(i) for i in x]))
recommendation.head()
```

```python
recommendation["tags"] = recommendation["tags"].apply(lambda x: x.lower())
```

Removal of Puctuation

```python
import string
def clean_text(text):
    return str(text).translate(str.maketrans('', '', string.punctuation)).strip()
recommendation["tags"] = recommendation["tags"].apply(clean_text)
```

```python
recommendation["tags"][0]
```

Stemming

```python
import nltk
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemer(text):
    lis = []
    for i in text.split():
        lis.append(ps.stem(i))
    return ' '.join(lis)
recommendation["tags"] = recommendation["tags"].apply(stemer)
```

### Word Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vector = TfidfVectorizer(max_features=5500,stop_words="english")
tfidf_wor = tfidf_vector.fit_transform(recommendation["tags"]).toarray()
```

```python
tfidf_vector.get_feature_names_out()
```

Calculating Similarity between Movies

```python
from sklearn.metrics.pairwise import cosine_similarity
similar = cosine_similarity(tfidf_wor)
```

```python
def recommend(movie):
    movie_index = recommendation[recommendation["title"]==movie].index[0]
    distances = similar[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(recommendation.iloc[i[0]].title)
recommend("Batman")
```

```python
import pickle 
pickle.dump(recommendation.to_dict(), open('Movies_dict.pkl', 'wb'))
```
